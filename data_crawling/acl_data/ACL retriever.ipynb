#### Automated Retrieval and Extraction of 2022 to 2024 Conference Papers from ACL Anthology and OpenReview 

This Python script automates the retrieval and extraction of academic papers from various conferences hosted on the ACL Anthology and OpenReview platforms. By specifying conference names and years, the script dynamically constructs URLs, scrapes paper details including title, authors, publication date, summary, and PDF link, and saves the collected data in JSON format. The script efficiently handles different HTML structures and provides progress tracking for large data sets.

#### Final code for retrieving papers from TACL 2023, part of the Gold Standard 

```python
import json
from bs4 import BeautifulSoup
import requests
import re
from tqdm import tqdm
from urllib.parse import urljoin

conference = 'tacl'
year = 2023

conference_url = f"https://aclanthology.org/events/{conference}-{year}/"

html_doc = requests.get(conference_url).text
soup = BeautifulSoup(html_doc, 'html.parser')

main_papers = soup.find_all('a', href=re.compile(rf"/{year}.{conference}-\d+.\d+/$"))

# no of papers to be found 
print(f"Found {len(main_papers)} papers.")

final = []

base_url = 'https://aclanthology.org/'

for paper in tqdm(main_papers, desc=f"{conference}{year}"):
    link = urljoin(base_url, paper['href'])
    # print the link to the papers
    print(f"Fetching paper: {link}")
    try:
        paper_html = requests.get(link).text
        tmp_soup = BeautifulSoup(paper_html, 'html.parser')
        
        title_elem = tmp_soup.find('h2', id='title')
        abstract_elem = tmp_soup.find('div', class_="card-body acl-abstract").find('span')
        authors_elem = tmp_soup.find('p', class_='lead')
        published_elem = tmp_soup.find('dt', text='Year:').find_next_sibling('dd')
        
        print(f"Title element: {title_elem}")
        print(f"Abstract element: {abstract_elem}")
        print(f"Authors element: {authors_elem}")
        print(f"Published element: {published_elem}")

        if title_elem and abstract_elem and authors_elem and published_elem:
            title = title_elem.get_text().strip()
            abstract = abstract_elem.get_text().strip()
            authors = [author.get_text().strip() for author in authors_elem.find_all('a')]
            published = published_elem.get_text().strip()
            pdf_link = tmp_soup.find('meta', {'name': 'citation_pdf_url'})['content']
            
            paper_info = {
                "title": title,
                "authors": authors,
                "published": published,
                "summary": abstract,
                "pdf_link": pdf_link
            }
            final.append(paper_info)
        else:
            print(f"Missing information for paper: {link}")
    except Exception as e:
        print(f'Error fetching paper: {link}')
        print(e)
print(f"Collected {len(final)} papers.")

json_path = f'{conference}{year}.json'
with open(json_path, 'w', encoding='utf-8') as jsonf:
    json.dump(final, jsonf, ensure_ascii=False, indent=4)
```

```python
import json
from bs4 import BeautifulSoup
import requests
import re
from tqdm import tqdm
from urllib.parse import urljoin

conferences = {
    "tacl2023": "https://aclanthology.org/events/tacl-2023/",
    "iclr2022": "https://openreview.net/group?id=ICLR.cc/2022/Conference",
    "acl2022": "https://aclanthology.org/events/acl-2022/",
    "naacl2022": "https://aclanthology.org/events/naacl-2022/",
    "emnlp2022": "https://aclanthology.org/events/emnlp-2022/",
    "aacl2022": "https://aclanthology.org/events/aacl-2022/",
    "iclr2023": "https://openreview.net/group?id=ICLR.cc/2023/Conference",
    "eacl2023": "https://aclanthology.org/events/eacl-2023/",
    "acl2023": "https://aclanthology.org/events/acl-2023/",
    "emnlp2023": "https://aclanthology.org/events/emnlp-2023/"
}

def fetch_papers(conference, year):
    conference_key = f"{conference.lower()}{year}"
    if conference_key not in conferences:
        print(f"Conference {conference} {year} is not in the list.")
        return []

    conference_url = conferences[conference_key]
    html_doc = requests.get(conference_url).text
    soup = BeautifulSoup(html_doc, 'html.parser')

    if "openreview.net" in conference_url:
        main_papers = soup.find_all('a', href=re.compile(r'/forum\?id=[^&]+$'))
    else:
        main_papers = soup.find_all('a', href=re.compile(rf"/{year}.{conference.lower()}-[a-z]+.\d+/$"))

    print(f"Found {len(main_papers)} papers for {conference} {year}.")

    final = []
    base_url = 'https://aclanthology.org/' if "aclanthology.org" in conference_url else 'https://openreview.net/'

    for paper in tqdm(main_papers, desc=f"{conference}{year}"):
        link = urljoin(base_url, paper['href'])
        print(f"Fetching paper: {link}")
        try:
            paper_html = requests.get(link).text
            tmp_soup = BeautifulSoup(paper_html, 'html.parser')
            
            if "aclanthology.org" in conference_url:
                title_elem = tmp_soup.find('h2', id='title')
                abstract_elem = tmp_soup.find('div', class_="card-body acl-abstract").find('span')
                authors_elem = tmp_soup.find('p', class_='lead')
                published_elem = tmp_soup.find('dt', text='Year:').find_next_sibling('dd')
                pdf_link_elem = tmp_soup.find('meta', {'name': 'citation_pdf_url'})

                if title_elem and abstract_elem and authors_elem and published_elem and pdf_link_elem:
                    title = title_elem.get_text().strip()
                    abstract = abstract_elem.get_text().strip()
                    authors = [author.get_text().strip() for author in authors_elem.find_all('a')]
                    published = published_elem.get_text().strip()
                    pdf_link = pdf_link_elem['content']
                    
                    paper_info = {
                        "title": title,
                        "authors": authors,
                        "published": published,
                        "summary": abstract,
                        "pdf_link": pdf_link
                    }
                    final.append(paper_info)
                else:
                    print(f"Missing information for paper: {link}")

            elif "openreview.net" in conference_url:
                title_elem = tmp_soup.find('h1', class_='title')
                abstract_elem = tmp_soup.find('span', class_="note-content-value")
                authors_elem = tmp_soup.find('span', class_="note-content-field", string="Authors:").find_next_sibling('span')
                published_elem = year 
                pdf_link_elem = tmp_soup.find('a', string='PDF')

                if title_elem and abstract_elem and authors_elem and pdf_link_elem:
                    title = title_elem.get_text().strip()
                    abstract = abstract_elem.get_text().strip()
                    authors = [author.strip() for author in authors_elem.get_text().split(',')]
                    pdf_link = urljoin(base_url, pdf_link_elem['href'])

                    paper_info = {
                        "title": title,
                        "authors": authors,
                        "published": published,
                        "summary": abstract,
                        "pdf_link": pdf_link
                    }
                    final.append(paper_info)
                else:
                    print(f"Missing information for paper: {link}")

        except Exception as e:
            print(f'Error fetching paper: {link}')
            print(e)

    print(f"Collected {len(final)} papers for {conference} {year}.")

    return final

def save_papers_to_json(conference, year):
    papers = fetch_papers(conference, year)
    json_path = f'{conference.lower()}{year}.json'
    with open(json_path, 'w', encoding='utf-8') as jsonf:
        json.dump(papers, jsonf, ensure_ascii=False, indent=4)
    print(f"Data successfully saved to {json_path}.")

# Example usage
for conference, year in [('ICLR', 2022), ('ACL', 2022), ('NAACL', 2022), ('EMNLP', 2022), ('AACL', 2022), ('ICLR', 2023), ('EACL', 2023), ('ACL', 2023), ('EMNLP', 2023)]:
    save_papers_to_json(conference, year)

    
```

#### Retrieving papers from NAACL 2024

```python
import json
from bs4 import BeautifulSoup
import requests
import re
from tqdm import tqdm
from urllib.parse import urljoin
import os

conference_url = "https://aclanthology.org/events/naacl-2024/"

def fetch_papers_naacl_2024():
    html_doc = requests.get(conference_url).text
    soup = BeautifulSoup(html_doc, 'html.parser')

    main_papers = soup.find_all('a', href=re.compile(r"/2024.naacl-[a-z]+.\d+/$"))
    print(f"Found {len(main_papers)} papers for NAACL 2024.")

    final = []
    base_url = 'https://aclanthology.org/'

    for paper in tqdm(main_papers, desc="NAACL2024"):
        link = urljoin(base_url, paper['href'])
        print(f"Fetching paper: {link}")
        try:
            paper_html = requests.get(link).text
            tmp_soup = BeautifulSoup(paper_html, 'html.parser')
            
            title_elem = tmp_soup.find('h2', id='title')
            abstract_elem = tmp_soup.find('div', class_="card-body acl-abstract").find('span')
            authors_elem = tmp_soup.find('p', class_='lead')
            published_elem = tmp_soup.find('dt', text='Year:').find_next_sibling('dd')
            pdf_link_elem = tmp_soup.find('meta', {'name': 'citation_pdf_url'})

            if title_elem and abstract_elem and authors_elem and published_elem and pdf_link_elem:
                title = title_elem.get_text().strip()
                abstract = abstract_elem.get_text().strip()
                authors = [author.get_text().strip() for author in authors_elem.find_all('a')]
                published = published_elem.get_text().strip()
                pdf_link = pdf_link_elem['content']
                
                paper_info = {
                    "title": title,
                    "authors": authors,
                    "published": published,
                    "summary": abstract,
                    "pdf_link": pdf_link
                }
                final.append(paper_info)
            else:
                print(f"Missing information for paper: {link}")

        except Exception as e:
            print(f'Error fetching paper: {link}')
            print(e)

    print(f"Collected {len(final)} papers for NAACL 2024.")

    return final

def save_papers_to_json_naacl_2024():
    papers = fetch_papers_naacl_2024()
    output_dir = os.path.expanduser('~/Master Thesis Draft/acl_data/')
    os.makedirs(output_dir, exist_ok=True)
    json_path = os.path.join(output_dir, 'naacl2024.json')
    with open(json_path, 'w', encoding='utf-8') as jsonf:
        json.dump(papers, jsonf, ensure_ascii=False, indent=4)
    print(f"Data successfully saved to {json_path}.")

# Example usage
save_papers_to_json_naacl_2024()
```
